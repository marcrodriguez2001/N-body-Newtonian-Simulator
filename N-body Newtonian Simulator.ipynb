## N-Body Newtonian Simulator

According to classical mechanics, a system of $N$ particles can be completely described by knowing the time evolution of the position of each of the $N$ particles composing it, in the three dimensions of a real space.

We will denote the position of the $k$-th particle with a three componenet vector $\vec{x}_k=(x_k^1, x_k^2, x_k^3)$, and the **trajectory** of this particles (its position in each time) with the curve (vector moving in time) $\vec{x}_k(t)=(x_k^1(t), x_k^2(t), x_k^3(t))$, where $t\in\mathbb{R}$ is the observation time. 

According to classical mechanics, each particle exerts an influence on each of the other particles, called a **force**. We will denote the force exerted by the $j$-th particle on the $k$-th particle as the three component vector $\vec{F}_{kj}(\vec{x}_k(t), \vec{x}_j(t))=(F_{kj}^1, F_{kj}^2, F_{kj}^3)$, which depends exclusively on the position of both particles at that particular time (in principle only on their relative position if the laws of physics are considered to be homogeneous and isotropic in space), and not explicitly on time (by the conservation of total energy).

Then, by the conservation of total momentum, there must exist at the same time a force exerted by the $k$-th particle on the $j$-th such that: $\vec{F}_{jk}(\vec{x}_j(t), \vec{x}_k(t))=-\vec{F}_{kj}(\vec{x}_k(t), \vec{x}_j(t))$ (also known as the action-reaction law). Finally, the sum of all the forces on a particle is defined as the time variation of the momentum of each particle (the velocity $\frac{d\vec{x}_k(t)}{dt}\equiv \vec{v}_k(t)$ times a particular weighting parameter of the particle called its mass $m_k(t)$):

$$
\frac{d}{dt}\Big(m_k(t)\frac{d\vec{x}_k(t)}{dt}  \Big) = \sum_{j=1\\ j\neq k}^N \vec{F}_{kj}(\vec{x}_k(t), \vec{x}_j(t)) \quad \text{for each particle k$\in$\{1,2,...,N\}}
$$

By assuming the mass of each particle is constant, this is also known as Newton's second law:

$$
 m_k\frac{d^2\vec{x}_k(t)}{dt^2}  = \sum_{j=1\\ j\neq k}^N \vec{F}_{kj}(\vec{x}_k(t), \vec{x}_j(t)) \quad \text{for each particle k$\in$\{1,2,...,N\}}
$$

If for example we have only three particles $\vec{x}_1,\vec{x}_2,\vec{x}_3$, we will have the system of differential equations:

$$
\begin{cases}
m_1\frac{d^2\vec{x}_1(t)}{dt^2}=\vec{F}_{12}(\vec{x}_1(t), \vec{x}_2(t)) +\vec{F}_{13}(\vec{x}_1(t), \vec{x}_3(t)) \\
m_2\frac{d^2\vec{x}_2(t)}{dt^2}=\vec{F}_{21}(\vec{x}_2(t), \vec{x}_1(t)) +\vec{F}_{23}(\vec{x}_2(t), \vec{x}_3(t)) 
\\
m_3\frac{d^2\vec{x}_3(t)}{dt^2}=\vec{F}_{31}(\vec{x}_3(t), \vec{x}_1(t)) +\vec{F}_{32}(\vec{x}_3(t), \vec{x}_2(t)) 
\end{cases}
$$


We can write each of these second order equations as two first order differential equations with $\frac{d\vec{x}_k(t)}{dt}\equiv \vec{v}_k(t)$:

$$
\begin{cases} m_k\frac{d\vec{v}_k(t)}{dt}  = \sum_{j=1\\ j\neq k}^N \vec{F}_{kj}(\vec{x}_k(t), \vec{x}_j(t)) \\
\frac{d\vec{v}_k(t)}{dt} = \vec{x}_k(t)
\end{cases}\quad \text{for each particle k$\in$\{1,2,...,N\}}
$$

Let us explain how to get the dynamics of the system if we know its initial conditions. We will get the trajectory of the system at times $t_0,t_0+\Delta t,t_o+2\Delta t,...$ for some small $\Delta t$.

If we know, at a initial time $t_0$, the position of the $N$ particles $\vec{x}_1(t_0),...,\vec{x}_N(t_0)$, we can get the total sum of force vectors affecting to each particle at the initial time $t_0$, by just evaluating the positions of the particles in the force functions:
$$
\textbf{(Step 1)}\quad \vec{F}_{Total\ on\ k}(t_0) := \sum_{j=1\\ j\neq k}^N \vec{F}_{kj}(\vec{x}_k(t_0), \vec{x}_j(t_0)) 
$$
Then, if we also know the velocity vector of the particles at that initial time $t_0$, $\vec{v}_1(t_0),...,\vec{v}_N(t_0)$ we can use the finite difference approximation for the derivative:

$$
\frac{df(t_0)}{dt}\simeq \frac{f(t_0+\Delta t)-f(t_0)}{\Delta t}
$$

in order to get the velocity vector at time $t+\Delta t$ from Newton's second law (with $\Delta t$ a small time increment):

$$
\frac{d\vec{v}_k(t_0)}{dt} = \frac{1}{m_k}\vec{F}_{Total\ on\ k}(t_0)\Longrightarrow \frac{\vec{v}_k(t_0+\Delta t)-\vec{v}_k(t_0)}{\Delta t}\simeq \frac{1}{m_k}\vec{F}_{Total\ on\ k}(t_0)\Longrightarrow
$$

$$
\textbf{(Step 2)}\quad \vec{v}_k(t_0+\Delta t) \simeq \vec{v}_k(t_0)+\frac{1}{m_k}\vec{F}_{Total\ on\ k}(t_0)\Delta t
$$

Finally, using the definition of the velocity vector we get the position of the particles at time $t_0+\Delta t$:

$$
\frac{d\vec{x}_k(t_0)}{dt} = \vec{v}_{k}(t_0)\Longrightarrow \frac{\vec{x}_k(t_0+\Delta t)-\vec{x}_k(t_0)}{\Delta t}\simeq \vec{v}_{k}(t_0)\Longrightarrow
$$

$$
\textbf{(Step 3)}\quad \vec{x}_k(t_0+\Delta t) \simeq \vec{x}_k(t_0)+\vec{v}_{k}(t_0)\Delta t
$$

Now that we know the positions of the $N$ particles at time $t_0+\Delta t$ we can again compute with Step 1 the total force on each particle at time $t_0+\Delta t$, with which we can compute the velocity vectors of the $N$ particles at $t_0+2\Delta t$ using Step 2, and then the positions of the $N$ particles at that new time using Step 3. Iterating this over and over, we can get the trajectories of the $N$ particles in time! 


Since the approximation for the derivative we are using is quite bad, we will need a very small $\Delta t$ for the solutions to be at least a bit stable, and certainly the particles will suddenly jump to infinity when errors start to build up, but until then, we will have a pretty cute simulator!

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import imageio
import os
```

**Definition of gravitational and Coulomb forces:**

Gravitational Force:
\begin{equation}
\vec{F_{12}} =G \frac{m_1 m_2}{r_{12}^{2}} \hat{r_{12}}=-G \frac{m_1 m_2}{r_{21}^{2}} \hat{r_{21}}
\end{equation}

Colulomb Force:
\begin{equation}
\vec{F}_{12} = k \frac{q_{1} q_{2}}{r_{12}^{2}} \hat{r}_{12}
\end{equation}

```python
def gravity_force(x1, x2, m1, m2,q1, q2, G=1.0):
    '''
    x1 and x2 are expected to be arrays of 3 elements (1 dimension) or 3x1 or 1x3 
    It returns a 3 element array back (the force vector)
    '''
    unit_vector = (x2-x1)/np.linalg.norm(x2-x1) # pointing from particle 1 to 2
    return (G*m1*m2/( np.linalg.norm(x2-x1)**2))*unit_vector

def coulomb_force(x1, x2, m1, m2, q1, q2, k=1.0):
    '''
    x1 and x2 are expected to be arrays of 3 elements (1 dimension) or 3x1 or 1x3 
    It returns a 3 element array back (the force vector)
    '''
    unit_vector = -(x2-x1)/np.linalg.norm(x2-x1) # pointing from particle 2 to 1 (here if charges equal force must be repulsive thus the minus sign)
    return (k*q1*q2/( np.linalg.norm(x2-x1)**2))*unit_vector
```

**Definition of the N-body Newtonian Simulator:**

```python
def run_n_body_simulator(N, positions_now, velocities_now, masses, charges, force_function, t0, tf, time_steps, limits, name):
    
    os.makedirs(f"./images_{name}", exist_ok=True)
    image_paths_list = []

    fig = plt.figure(figsize=(10, 10))

    # Compute the times for simulation steps
    time_list = np.linspace(start=t0, stop=tf, num=time_steps)
    dt = time_list[1] - time_list[0]  # Time Step

    # Compute positions and velocities lists
    positions = np.zeros((time_steps+1, N, 3))
    velocities = np.zeros((time_steps+1, N, 3))
    
    # Initialize positions and velocities
    positions[0] = positions_now
    velocities[0] = velocities_now

    # Verlet algorithm
    for step, time in enumerate(time_list):
        
        forces = np.zeros((N, 3))
        
        # Step 1: compute the total force on each particle
        for i in range(N):
            for j in range(N):
                if j != i: # Skip calculating force with itself
                    forces[i, :] = forces[i, :]+ force_function(positions[step, i], positions[step, j], masses[i], masses[j], charges[i], charges[j])
        
        # Step 2: update velocities using forces and masses
        velocities[step+1] = velocities[step]+ (forces / np.array(masses)[:, np.newaxis]) * dt
        
        # Step 3: update positions using the Verlet algorithm
        positions[step+1] = positions[step] + velocities[step] * dt
        
        # Plot the particles
        image_path= f"./images_{name}/NBody_{step}.png"
        
        plot(fig, image_path, positions, velocities, time, N, step,limits)
        
        image_paths_list.append(image_path)

        
        print(f'Step: {step}/{len(time_list)}')
    
    gif_images = [imageio.imread(image_path) for image_path in image_paths_list]
    
    imageio.mimsave(f'{N}Body_Simulation_{name}.gif', gif_images, fps=10)
    
    print('Gif Generated')
```

**Definition of the plot particles function:**

```python
def plot(fig, image_path, positions, velocities, time, N, step, limits):
    
    plt.clf()
    
    figure = fig.add_subplot(111, projection='3d')
    
    color = matplotlib.cm.get_cmap('hsv')
    
    # Plot particles
    figure.scatter3D(positions[step,:,0], positions[step,:,1], positions[step,:,2], c=color(np.arange(N)/N), s=100)
    
    # Plot trajectories by plotting only the current position of each particle
    for i in range(positions.shape[1]): 
        figure.plot3D(positions[:step, i, 0], positions[:step, i, 1], positions[:step, i, 2], '--', c=color(i/N))
    
    # Graph options
    xmin= limits['xmin']
    xmax= limits['xmax'] 
    ymin= limits['ymin']
    ymax= limits['ymax']
    zmin= limits['zmin']
    zmax= limits['zmax']
    
    figure.set_xlabel('X')
    figure.set_xlim((xmin, xmax))
    
    figure.set_ylabel('Y')
    figure.set_ylim((ymin, ymax))

    figure.set_zlabel('Z')
    figure.set_zlim((zmin, zmax))
    
    figure.set_title(f"{N} Body simulation \nTime:{time:4.3}")
    
    plt.tight_layout
    plt.savefig(image_path, dpi=200)
```

**Examples of initial conditions:**

Example of 2 bodies with same mass Gravitational attraction:

```python
N = 2  # Number of particles
positions_now = np.array([[0, 0, 0], [1, 1, 1]])  # Initial positions of particles
velocities_now = np.array([[0, 0, 0], [0, 0, 0]])  # Initial velocities of particles
masses = [1, 1]  # Masses of particles
charges = [0, 0] # Charges of particles

# Type of force

force_function=gravity_force

# Set up simulation parameters
t0 = 0
tf = 1.5
time_steps = 100
limits = {'xmin': -3, 'xmax': 3, 'ymin': -3, 'ymax': 3, 'zmin': -3, 'zmax': 3}
name="2_body_gravitational_force"

# Run the N-body simulator
run_n_body_simulator(N, positions_now, velocities_now, masses,charges, force_function, t0, tf, time_steps, limits, name)
```

3 body problem is a challenge in physics that deals with predicting the movements of three objects influenced by gravity. Analytically solving this problem is impossible due to its complexity, as there's no general formula to predict their exact positions over time. However, simulation methods can be used to approximate their behavior by performing numerical calculations. Example of 3 body problem:

```python
run_n_body_simulator(N=3, positions_now=np.array([[0.97000436,-0.24308753,0], 
                                                  [-0.97000436,+0.24308753,0], 
                                                  [0,0,0]
                                                 ]), 
                    velocities_now=np.array([[0.93240737/2, 0.86473146/2, 0], 
                                             [0.93240737/2, 0.86473146/2, 0],
                                             [-0.93240737, -0.86473146, 0]]),
                    masses=[1, 1, 1,], 
                    charges =[0, 0, 0], 
                    force_function =gravity_force,
                    t0=0, tf=10, time_steps=1000,
                    limits={'xmin':-1, 'xmax':1, 'ymin':-1, 'ymax':1, 'zmin':-1, 'zmax':1},
                     name="GravitySimulation"
                    )
```
